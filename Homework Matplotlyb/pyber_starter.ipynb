```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv("data/city_data.csv")
ride_df = pd.read_csv("data/ride_data.csv")
city_df.head()
ride_df.head()

# Combine the data into a single dataset
merge_df=pd.merge(city_df,ride_df,on="city")

# Display the data table for preview
merge_df.head()
```

```python
# Display the types of cities
merge_df.type.unique()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x y coordinates for 3 cities
urban_df = merge_df.loc[merge_df['type'] == "Urban"]
suburban_df = merge_df.loc[merge_df['type'] == "Suburban"]
rural_df = merge_df.loc[merge_df['type'] == "Rural"]
urban_group = urban_df.groupby(['city'])
suburban_group = suburban_df.groupby(['city'])
rural_group = rural_df.groupby(['city'])

urban_driver_count = urban_group.ride_id.count()
suburban_driver_count = suburban_group.ride_id.count()
rural_driver_count = rural_group.ride_id.count()
urban_total_fares = urban_group.fare.sum()
suburban_total_fares = suburban_group.fare.sum()
rural_total_fares = rural_group.fare.sum()

urban_avg_fare = urban_total_fares / urban_driver_count
suburban_avg_fare = suburban_total_fares / suburban_driver_count
rural_avg_fare = rural_total_fares / rural_driver_count

```

```python
# Build the scatter plots for each city types ##Recuerda que son los df del inicio ### Recuerda 
print(city_df)
city_new_df = city_df.sort_index()


urban_df = city_new_df.loc[city_new_df['type'] == "Urban"]
suburban_df = city_new_df.loc[city_new_df['type'] == "Suburban"]
rural_df = city_new_df.loc[city_new_df['type'] == "Rural"]

urban_bubble = urban_df['driver_count'] * 10
suburban_bubble = suburban_df['driver_count'] * 10
rural_bubble = rural_df['driver_count'] * 10

urban_handle = plt.scatter(urban_driver_count, urban_avg_fare, marker="o", c='orange', edgecolors="black", s=urban_bubble, alpha=0.75, label="Urban")
suburban_handle = plt.scatter(suburban_driver_count, suburban_avg_fare, marker="o", c='lightskyblue', edgecolors="black", s=suburban_bubble, alpha=0.75, label="Suburban")
rural_handle = plt.scatter(rural_driver_count, rural_avg_fare, marker="o", c='gold', edgecolors="black", s=rural_bubble, alpha=0.75, label="Rural")
```

```python

# Other graph properties
plt.xlim(0, 40)
plt.ylim(10, 45)
plt.grid(axis='both', alpha=0.5)
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
```
